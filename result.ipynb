```python
def cross_entropy_loss(logits, labels):
    return F.cross_entropy(logits.view(-1, logits.size(-1)), labels.view(-1))

def mutual_information(logits, labels):
    # 
    probs = F.softmax(logits, dim=-1)
    
    #  H(y)
    H_y = -torch.mean(torch.log(probs + 1e-9))

    #  H(y|x)
    log_probs = F.log_softmax(logits, dim=-1)
    H_y_given_x = F.nll_loss(log_probs.view(-1, logits.size(-1)), labels.view(-1))
    
    #  I(x; y) = H(y) - H(y|x)
    I_x_y = H_y - H_y_given_x
    return I_x_y


def custom_loss(logits, labels):
    # 
    ce_loss = cross_entropy_loss(logits, labels)
    
    # 
    mi = mutual_information(logits, labels)
    
    #  + 
    loss = ce_loss + 0.8 * mi
    return loss
```
